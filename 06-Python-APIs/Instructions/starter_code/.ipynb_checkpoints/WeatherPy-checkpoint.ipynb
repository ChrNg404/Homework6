```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
print(cities[0])
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# So let's set up the URL so we can start doing API calls
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial" #"[metric, imperial]"

#query_url = f"{url}appid={api_key}&q={cities[0]}&units={units}"
#weather_response = requests.get(query_url)
#weather_json=weather_response.json()
#pprint(weather_json)

# Next we make some empty lists to hold everything in. 

Temperature =[]
MaxTemp = []
Humidity = []
Cloudiness = []
Wind_Speed = []
lat = []

# After that we set up a loop to pull data on each of the cities from the randomized list
# We'll use Try here because some cities aren't in the database

for index, city in enumerate(cities):
    try:
        query_url = f"{url}appid={api_key}&q={city}&units={units}"
        response = requests.get(query_url).json()
        Temperature.append(response['main']['temp'])
        MaxTemp.append(response['main']['temp_max'])
        Humidity.append(response['main']['humidity'])
        Cloudiness.append(response['clouds']['all'])
        Wind_Speed.append(response['wind']['speed'])
        lat.append(response['coord']['lat'])
        #lat_lngs.append(response['coord'])
        #cities.append(response['name'])
        print(f"City #{index}: {city} is being processed.")
        #pprint(response)
    except:
        Temperature.append(np.nan)
        MaxTemp.append(np.nan)
        Humidity.append(np.nan)
        Cloudiness.append(np.nan)
        Wind_Speed.append(np.nan)
        lat.append(np.nan)
        print(f"City #{index}: {city} could not be found.")
        
    #weather_response = requests.get(query_url)
    #weather_json = weather_response.json()
    #pprint(weather_json)
```

```python
# Here's just the first city in the randomized list for reference of what data we will get
query_url = f"{url}appid={api_key}&q={cities[0]}&units={units}"
weather_response = requests.get(query_url)
weather_json=weather_response.json()
pprint(weather_json)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# So let's turn the lists into a dictionary
Weather_Data = {
    "Names":cities,
    "Temperature": Temperature,
    "Maximum Temperature": MaxTemp,
    "Humidity" : Humidity,
    "Cloudiness" : Cloudiness,
    "Wind Speed" : Wind_Speed,
    "Lat" : lat
}

# Then from a dictionary we're going to turn it into a dataframe

weather_df = pd.DataFrame.from_dict(Weather_Data)
weather_df.head()
```

```python
# Let's do a quick count to see how many cells of data we have
weather_df.count()
```

```python
# Then we're going to drop the cells that don't have data in them, and we'll call the variable so we can make sure the missing data lines were dropped
weather_df = weather_df.dropna(how='any')
weather_df
```
